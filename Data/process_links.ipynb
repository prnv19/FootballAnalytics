```python
import pandas as pd
import os
```

```python
cwd = os.getcwd()
links_df = pd.read_csv("fbref_to_tm_mapping.csv", sep=",", encoding='latin-1')

#Convert columns to list
fbref_links = links_df['UrlFBref'].tolist()
names = [" ".join(i.split("/")[-1].split("-")) for i in fbref_links]
links_df['Name'] = names
links_df.drop(columns=['UrlFBref','PlayerFBref','TmPos'], inplace=True)
links_df.rename(columns={"UrlTmarkt": "Url"}, inplace=True)

#Get ID from player link
links_df['ID'] = [i.split("/")[-1] for i in links_df['Url']]
links_df.to_csv(os.path.join(cwd, "player_links.csv"), index=False) 

links_df
```
