## Gaussian Naive Bayes

```python
import pandas as pd
import os

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils import resample

import warnings
warnings.filterwarnings("ignore")
```

## Dataset Creation

```python
parent_dir = os.path.dirname(os.getcwd())
data = pd.read_csv(os.path.join(parent_dir, 'Data', 'Classification_data.csv'))
data
```

```python
numeric_columns = ['Age', '20YC', '20YC2', '20RC', '20G', '20A', '20MP', '20AP', '21YC', '21YC2', '21RC', '21G', '21A', '21MP', '21AP', '22YC', '22YC2', '22RC', '22G', '22A', '22MP', '22AP', '23YC', '23YC2', '23RC', '23G', '23A', '23MP', '23AP', '24YC', '24YC2', '24RC', '24G', '24A', '24MP', '24AP', '25YC', '25YC2', '25RC', '25G', '25A', '25MP', '25AP', 'TotalCups']
target_size = 1000
train_size = 0.8

data_GNB = data[numeric_columns+ ['ValueCategory']]

# Split into features and target variable
X = data_GNB.drop(columns=['ValueCategory'])
y = data_GNB['ValueCategory']

# Separate classes
X_0, X_1, X_2 = X[y == 0], X[y == 1], X[y == 2]
y_0, y_1, y_2 = y[y == 0], y[y == 1], y[y == 2]

# Resample each class to match the target size
X_0_balanced, y_0_balanced = resample(X_0, y_0, n_samples=target_size, random_state=1911, replace=True)
X_1_balanced, y_1_balanced = resample(X_1, y_1, n_samples=target_size, random_state=1911, replace=True)
X_2_balanced, y_2_balanced = resample(X_2, y_2, n_samples=target_size, random_state=1911, replace=True)

# Combine balanced samples
X_balanced = pd.concat([X_0_balanced, X_1_balanced, X_2_balanced], axis=0)
y_balanced = pd.concat([y_0_balanced, y_1_balanced, y_2_balanced], axis=0)

# Perform train-test split again after removing outliers
X_train_GNB, X_test_GNB, y_train_GNB, y_test_GNB = train_test_split(
    X_balanced, y_balanced, train_size=train_size, random_state=1911, shuffle=True
)

# X_train_GNB, X_test_GNB, y_train_GNB, y_test_GNB = train_test_split(X, y, train_size=train_size, random_state=1911, stratify=y, shuffle=True)

# Standardize numerical features
scaler = StandardScaler()
X_train_GNB[numeric_columns] = scaler.fit_transform(X_train_GNB[numeric_columns])
X_test_GNB[numeric_columns] = scaler.transform(X_test_GNB[numeric_columns])


X_train_GNB.reset_index(drop=True, inplace=True)
X_test_GNB.reset_index(drop=True, inplace=True)
y_train_GNB.reset_index(drop=True, inplace=True)
y_test_GNB.reset_index(drop=True, inplace=True)

X_train_GNB.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'GNB' ,'X_train_GNB.csv'), index=False)
X_test_GNB.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'GNB' ,'X_test_GNB.csv'), index=False)
y_train_GNB.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'GNB' ,'y_train_GNB.csv'), index=False)
y_test_GNB.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'GNB' ,'y_test_GNB.csv'), index=False)

X_train_GNB.shape, X_test_GNB.shape, y_train_GNB.shape, y_test_GNB.shape
```

```python
from sklearn.naive_bayes import GaussianNB

# Train Naïve Bayes classifier
nb_model = GaussianNB()
nb_model.fit(X_train_GNB, y_train_GNB)

# Predictions
y_pred_GNB = nb_model.predict(X_test_GNB)

# Evaluation
print("Gaussian Naive Bayes Accuracy:", accuracy_score(y_test_GNB, y_pred_GNB))
print("\nClassification Report:")
print(classification_report(y_test_GNB, y_pred_GNB))
print("\nConfusion Matrix:")
cm = confusion_matrix(y_test_GNB, y_pred_GNB)
ConfusionMatrixDisplay(confusion_matrix=cm).plot(cmap='Blues')
```
