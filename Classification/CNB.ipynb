## Categorical Naieve Bayes

```python
import os
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils import resample

import warnings
warnings.filterwarnings("ignore")
```

## Dataset Creation

```python
parent_dir = os.path.dirname(os.getcwd())
data = pd.read_csv(os.path.join(parent_dir, 'Data', 'Classification_data.csv'))
data
```

```python
# Define categorical and numerical columns 
categorical_columns = ['Foot', 'Position', 'OtherPosition', 'National', 'Club_name', 'ContractOption', 'Outfitter']
target_size = 1000
train_size = 0.8

# Keep only categorical columns and target variable
data_CNB = data[categorical_columns+ ['ValueCategory']]

# Apply Label Encoding to categorical features
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    data_CNB[col] = le.fit_transform(data_CNB[col].astype(str))  
    label_encoders[col] = le

# Split into features and target variable
X = data_CNB.drop(columns=['ValueCategory'])
y = data_CNB['ValueCategory']

# Separate classes
X_0, X_1, X_2 = X[y == 0], X[y == 1], X[y == 2]
y_0, y_1, y_2 = y[y == 0], y[y == 1], y[y == 2]

# Resample each class to match the target size
X_0_balanced, y_0_balanced = resample(X_0, y_0, n_samples=target_size, random_state=1911, replace=True)
X_1_balanced, y_1_balanced = resample(X_1, y_1, n_samples=target_size, random_state=1911, replace=True)
X_2_balanced, y_2_balanced = resample(X_2, y_2, n_samples=target_size, random_state=1911, replace=True)

# Combine balanced samples
X_balanced = pd.concat([X_0_balanced, X_1_balanced, X_2_balanced], axis=0)
y_balanced = pd.concat([y_0_balanced, y_1_balanced, y_2_balanced], axis=0)

# Perform train-test split again after removing outliers
X_train_CNB, X_test_CNB, y_train_CNB, y_test_CNB = train_test_split(
    X_balanced, y_balanced, train_size=train_size, random_state=1911, shuffle=True
)

# Find the maximum value in each column of the training data
max_values = {}
for col in X_train_CNB.columns:
    max_values[col] = X_train_CNB[col].max()

# Check and cap values in test data to ensure they don't exceed the training data range
for col in X_test_CNB.columns:
    # Identify values exceeding the training range
    mask = X_test_CNB[col] > max_values[col]
    if mask.sum() > 0:
        print(f"Column {col} has {mask.sum()} values outside training range")
        # Replace those values with the maximum from training
        X_test_CNB.loc[mask, col] = max_values[col]
        

X_train_CNB.reset_index(drop=True, inplace=True)
X_test_CNB.reset_index(drop=True, inplace=True)
y_train_CNB.reset_index(drop=True, inplace=True)
y_test_CNB.reset_index(drop=True, inplace=True)

# X_train_CNB, X_test_CNB, y_train_CNB, y_test_CNB = train_test_split(X, y, train_size=train_size, random_state=1911, stratify=y, shuffle=True)


X_train_CNB.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'CNB' ,'X_train_CNB.csv'), index=False)
X_test_CNB.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'CNB' ,'X_test_CNB.csv'), index=False)
y_train_CNB.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'CNB' ,'y_train_CNB.csv'), index=False)
y_test_CNB.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'CNB' ,'y_test_CNB.csv'), index=False)

X_train_CNB.shape, X_test_CNB.shape, y_train_CNB.shape, y_test_CNB.shape
```

```python
from sklearn.naive_bayes import CategoricalNB

# Train Naive Bayes classifier
nb_model = CategoricalNB()
nb_model.fit(X_train_CNB, y_train_CNB)

# Predictions
y_pred_CNB = nb_model.predict(X_test_CNB)

# Evaluation
print("Categorical Naive Bayes Accuracy:", accuracy_score(y_test_CNB, y_pred_CNB))
print("\nClassification Report:")
print(classification_report(y_test_CNB, y_pred_CNB))
print("\nConfusion Matrix:")
cm = confusion_matrix(y_test_CNB, y_pred_CNB)
ConfusionMatrixDisplay(confusion_matrix=cm).plot(cmap='Blues')
```
