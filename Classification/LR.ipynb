## Logistic Regression

```python
import pandas as pd
import os
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils import resample

import warnings
warnings.filterwarnings("ignore")
```

## Dataset Creation

```python
parent_dir = os.path.dirname(os.getcwd())
data = pd.read_csv(os.path.join(parent_dir, 'Data', 'Classification_data.csv'))
data
```

```python
# Define categorical and numerical columns 
numeric_columns = ['Age', '20YC', '20YC2', '20RC', '20G', '20A', '20MP', '20AP', '21YC', '21YC2', '21RC', '21G', '21A', '21MP', '21AP', '22YC', '22YC2', '22RC', '22G', '22A', '22MP', '22AP', '23YC', '23YC2', '23RC', '23G', '23A', '23MP', '23AP', '24YC', '24YC2', '24RC', '24G', '24A', '24MP', '24AP', '25YC', '25YC2', '25RC', '25G', '25A', '25MP', '25AP', 'TotalCups']

target_size = 1000
train_size = 0.8

data = data[numeric_columns+ ['ValueCategory']]

# Split into features and target variable
X = data.drop(columns=['ValueCategory'])
y = data['ValueCategory']

# Separate classes
X_0, X_1 = X[y == 0], X[y == 1]
y_0, y_1 = y[y == 0], y[y == 1]


# Resample each class to match the target size
X_0_balanced, y_0_balanced = resample(X_0, y_0, n_samples=target_size, random_state=1911, replace=True)
X_1_balanced, y_1_balanced = resample(X_1, y_1, n_samples=target_size, random_state=1911, replace=True)

# Combine balanced samples
X_balanced = pd.concat([X_0_balanced, X_1_balanced], axis=0)
y_balanced = pd.concat([y_0_balanced, y_1_balanced], axis=0)
     
# Perform train-test split again after removing outliers
X_train, X_test, y_train, y_test = train_test_split(
    X_balanced, y_balanced, train_size=train_size, random_state=1911, shuffle=True
)

# X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_size, random_state=1911, stratify=y, shuffle=True)

float_columns = X_train.select_dtypes(include=['float64']).columns
X_train[float_columns] = np.floor(X_train[float_columns]).astype(int)
X_test[float_columns] = np.floor(X_test[float_columns]).astype(int)


X_train.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
y_train.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)

X_train.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'LR' ,'X_train_LR.csv'), index=False)
X_test.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'LR' ,'X_test_LR.csv'), index=False)
y_train.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'LR' ,'y_train_LR.csv'), index=False)
y_test.to_csv(os.path.join(parent_dir, 'Data', 'ModelData', 'LR' ,'y_test_LR.csv'), index=False)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

# Create and train the model
lr_model = LogisticRegression(random_state=19, max_iter=1000)
lr_model.fit(X_train, y_train)
# Make predictions
y_pred_lr = lr_model.predict(X_test)

# Evaluation
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr))
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, y_pred_lr)).plot(cmap='Blues')
```

```python
from sklearn.naive_bayes import MultinomialNB

# Train Naive Bayes classifier
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)

# Predictions
y_pred_MNB = nb_model.predict(X_test)

# Evaluation
print("Multinomial Naive Bayes Accuracy:", accuracy_score(y_test, y_pred_MNB))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_MNB))
print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred_MNB)
ConfusionMatrixDisplay(confusion_matrix=cm).plot(cmap='Blues')
```
